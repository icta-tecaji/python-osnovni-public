````python
import requests

url = "https://podatki.gov.si/dataset/ada88e06-14a2-49c4-8748-3311822e3585/resource/eb8b25ea-5c00-4817-a670-26e1023677c6/download/seznampraznikovindelaprostihdni20002030.csv"

response = requests.get(url)
response.encoding = "utf-8"

data = response.text
#print(data)

rezultat = {}
for vrstica in data.split("\r\n")[1:]:
    #print(vrstica.split(";"))
    leto = vrstica.split(";")[-1]
    #print(leto)
    if leto == "2022":
        #print(vrstica)
        dan = vrstica.split(";")[2]
        #print(dan)
        
        if dan in rezultat.keys():
            rezultat[dan] += 1
        else:
            rezultat[dan] = 1
            
            
print(rezultat)
````

````python
import requests

my_portfolio = ["bitcoin", "ethereum", "cardano", "polkadot", "secret"]

for coin in my_portfolio:
    url = f"https://api.coingecko.com/api/v3/simple/price?ids={coin}&vs_currencies=eur"
    r = requests.get(url)
    data = r.json()
    print(f"Cena {coin} je {data[coin]['eur']} €")
````

````python
import requests

my_portfolio = ["bitcoin", "ethereum", "cardano", "polkadot", "secret"]

for coin in my_portfolio:
    url = f"https://api.coingecko.com/api/v3/coins/{coin}/market_chart?vs_currency=eur&days=3&interval=daily"
    r = requests.get(url)
    
    data = r.json()
    
    for i in range(3):
        print(f'{coin}, {data["prices"][i][1]:.2f}€, \t {data["market_caps"][i][1]:.2f}')
    print()
    
````

````python
import requests

url = "https://opendata.si/promet/bicikelj/list/"

r = requests.get(url)
data = r.json()

#print(data)

free_bike = False
free_park = False
for key, station in data["markers"].items():
    #print(station["name"])
    if station["name"] == "TRG MDB":
        print(station["station"]["available"])
        if int(station["station"]["available"]) >= 1:
            free_bike = True
        
    if station["name"] == "STARA CERKEV":
        print(station["station"]["free"])
        if int(station["station"]["free"]) >= 1:
            free_park = True
            
if free_bike and free_park:
    print("Lahko greš z Bicikelj")
else:
    print("Ne moreš z bicikelj")
````
